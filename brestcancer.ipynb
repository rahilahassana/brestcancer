```python
import pandas as pd
import io
data=pd.read_csv(io.StringIO(uploaded['data.csv'].decode('utf-8')))
data.head(5)
```

```python
#seaborn is graphical visualisation library
import seaborn as sns
ax=sns.countplot(data['diagnosis'], label='Count')
B,M=data['diagnosis'].value_counts()
print("Benign",B)
print("Malignant",M)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
del data['Unnamed: 32']
```

```python
x=data.iloc[:,2:].values
y=data.iloc[:,1].values

#encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_x_1 = LabelEncoder()
y=labelencoder_x_1.fit_transform(y)

#splitting the data into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=0)
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test= sc.transform(x_test)




```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#adding the input and first hidden layer
classifier = Sequential()
classifier.add(Dense(16, kernel_initializer='uniform', activation='relu',input_dim=30))

#adding the second hidden layer
classifier.add(Dense(16, kernel_initializer='uniform', activation='relu'))

#adding the outut layer
classifier.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
  classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(x_train, y_train, batch_size=100, epochs=200)
```

```python
x_test
```

```python
#predecting the test set results
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm,annot=True)
plt.savefig('h.png')
```
